# Parse GO DB data

```python
import pandas as pd
import xml.etree.ElementTree as ET

def iter_records(records):
    temp_dict = []   
    for record in records:
        # temporary dictionary to hold values        
        
        print("====================================")
        print("====================================")
        try:
            print((record[0][1].text)) # complex name 
        except:
             print((record[0][0].text))
            
        print(record[1].text)                       # protien count 
        print(record[2][2].text)                    # expected        
        print(record[2][4].text)                    # pValue        
        print(record[2][5].text)                    # fold_enrichment        
        print(record[2][6].text)                    # fdr
        
        
        proteinlist=[]
        
        try:
            for i in record[2][7]:
                print(i.text)
                proteinlist.append(i.text) 
        except:
            print("Error Reading proeint list")
        
        
        try:
            temp_dict.append([record[0][1].text, int(record[1].text),record[2][2].text,record[2][4].text,
                          float(record[2][5].text),float(record[2][6].text),proteinlist]   )
        except:
             temp_dict.append([record[0][0].text,int(record[1].text),record[2][2].text,record[2][4].text,
                          float(record[2][5].text),float(record[2][6].text),proteinlist]   )
                
                
        
        print("====================================")
        print("====================================")
    return temp_dict
 
        
def read_xml(xmlFileName):
    with open(xmlFileName, 'r') as xml_file:
        # read the data and store it as a tree
        tree = ET.parse(xml_file)

        # get the root of the tree
        root = tree.getroot()
         
        # return the DataFrame
        return pd.DataFrame(list(iter_records(root)))
```

```python
utmb_data= pd.read_csv("compare.csv")[["Portien","rateconst"]]
utmb_list= pd.read_csv("Protein_List.utmb.0311_2022.csv")[["Entry","Entry name","Gene names"]]
utmb_data=utmb_data.merge(utmb_list,left_on=["Portien"],right_on=["Entry name"])[["Entry","Entry name",
                                                                                  "rateconst","Gene names"]]
utmb_data=utmb_data.reset_index(drop=True)
```

# Merge the data

```python
#search
found_list={}

for i in range(utmb_data.shape[0]): 
    
    for j in range(parsedData.shape[0]):
        
        temp=list(parsedData[ "protien_list"][j])
        temp=[x.strip() for x in temp]
        
        if((utmb_data["Entry"][i]).strip() in temp):
            
            
            if( (parsedData["complex_name"][j]) in found_list.keys()):
                
                found_list[parsedData["complex_name"][j]]["protiens"].append(utmb_data["Entry"][i])
                found_list[parsedData["complex_name"][j]]["protiens_name"].append(utmb_data["Entry name"][i])
                found_list[parsedData["complex_name"][j]]["k"].append(float(utmb_data["rateconst"][i]))
                
            else:
                found_list[(parsedData["complex_name"][j])]={"complex_name":parsedData["complex_name"][j],
                                                  "protiens":[utmb_data["Entry"][i]],
                                                  "protiens_name":[utmb_data["Entry name"][i]],
                                                  "k":[ float(utmb_data["rateconst"][i])]}

        
```

```python
df=pd.DataFrame(found_list)
df=df.transpose()
df=df.reset_index()
print(df.shape)
df.head(5)
```
