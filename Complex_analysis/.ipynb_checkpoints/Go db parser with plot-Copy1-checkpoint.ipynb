# Parse GO DB data

```python
import pandas as pd
import xml.etree.ElementTree as ET

def iter_records(records):
    temp_dict = []   
    for record in records:
        # temporary dictionary to hold values        
        
        print("====================================")
        print("====================================")
        try:
            print((record[0][1].text)) # complex name 
        except:
             print((record[0][0].text))
            
        print(record[1].text)                       # protien count 
        print(record[2][2].text)                    # expected        
        print(record[2][4].text)                    # pValue        
        print(record[2][5].text)                    # fold_enrichment        
        print(record[2][6].text)                    # fdr
        
        
        proteinlist=[]
        
        try:
            for i in record[2][7]:
                print(i.text)
                proteinlist.append(i.text) 
        except:
            print("Error Reading proeint list")
        
        
        try:
            temp_dict.append([record[0][1].text, int(record[1].text),record[2][2].text,record[2][4].text,
                          float(record[2][5].text),float(record[2][6].text),proteinlist]   )
        except:
             temp_dict.append([record[0][0].text,int(record[1].text),record[2][2].text,record[2][4].text,
                          float(record[2][5].text),float(record[2][6].text),proteinlist]   )
                
                
        
        print("====================================")
        print("====================================")
    return temp_dict
 
        
def read_xml(xmlFileName):
    with open(xmlFileName, 'r') as xml_file:
        # read the data and store it as a tree
        tree = ET.parse(xml_file)

        # get the root of the tree
        root = tree.getroot()
         
        # return the DataFrame
        return pd.DataFrame(list(iter_records(root)))
```

```python
parsedData=read_xml("C:/Workplace/Python/d2omeplusAnalysis/d2omePlusAnalysis/GO_Database_UTMB_mouse_liver/analysis - Copy.xml")
```

```python
parsedData.columns=['complex_name','protien_count','expected','pValue','fold_enrichment','fdr','protien_list']
parsedData.pValue=[float(x) for x in parsedData.pValue]
parsedData.to_csv('GODBB_raw_data.csv', index=False)
print(parsedData.shape)
parsedData.head()
```

```python
parsedData=parsedData[(parsedData.protien_count>=5)& (parsedData.fold_enrichment>1)& (parsedData.pValue<0.001)]
parsedData=parsedData.reset_index(drop=True)
print(parsedData.shape)
parsedData.head()
```

# Read utmb data

```python
# utmb_data= pd.read_csv("compare.csv")[["Portien","rateconst"]]
utmb_data= pd.read_csv("Analyzed_Proteins_and_Their_Rates.csv")[["Proteins"," RateConstant"]]
utmb_data.columns=["protiens","rateconst"]

utmb_list= pd.read_csv("Protein_List.utmb.0518_2022.csv")[["Entry","Entry name","Gene names"]]
utmb_data=utmb_data.merge(utmb_list,left_on=["protiens"],right_on=["Entry name"])[["Entry","Entry name",
                                                                                  "rateconst","Gene names"]]
utmb_data=utmb_data.reset_index(drop=True)
```

# Merge the data

```python
#search
found_list={}

for i in range(utmb_data.shape[0]): 
    
    for j in range(parsedData.shape[0]):
        
        temp=list(parsedData[ "protien_list"][j])
        temp=[x.strip() for x in temp]
        
        if((utmb_data["Entry"][i]).strip() in temp):
            
            
            if( (parsedData["complex_name"][j]) in found_list.keys()):
                
                found_list[parsedData["complex_name"][j]]["protiens"].append(utmb_data["Entry"][i])
                found_list[parsedData["complex_name"][j]]["protiens_name"].append(utmb_data["Entry name"][i])
                found_list[parsedData["complex_name"][j]]["k"].append(float(utmb_data["rateconst"][i]))
                
            else:
                found_list[(parsedData["complex_name"][j])]={"complex_name":parsedData["complex_name"][j],
                                                  "protiens":[utmb_data["Entry"][i]],
                                                  "protiens_name":[utmb_data["Entry name"][i]],
                                                  "k":[ float(utmb_data["rateconst"][i])]}

        
```

```python
def prepare_data(found_list, droplist):
    df=pd.DataFrame(found_list)
    df=df.transpose()
    print(df.shape)
    df=df.drop(index = droplist)#'late endosome membrane',

    df["pc"]=[len(k) for k in df.k]
    df=df[df.pc<100]
    df=df[df.pc>=5]
    df=df.sort_values(by=['pc'])
    df=df.reset_index(drop=True)
    print(df.shape)
    
    return df
droplist=['protein-lipid complex','blood microparticle',"endocytic vesicle membrane",
                        'lipoprotein particle','high-density lipoprotein particle','plasma lipoprotein particle',
                       'triglyceride-rich plasma lipoprotein particle',
                       'very-low-density lipoprotein particle','secondary lysosome','phagocytic vesicle',
                       'phagocytic vesicle membrane','U5 snRNP','MHC class I peptide loading complex',
                       'extrinsic component of organelle membrane','microbody lumen','peroxisomal matrix']

for i in ['protein-lipid complex', 'blood microparticle', 'endocytic vesicle membrane', 'lipoprotein particle', 'high-density lipoprotein particle', 'plasma lipoprotein particle', 'triglyceride-rich plasma lipoprotein particle', 'very-low-density lipoprotein particle', 'secondary lysosome', 'phagocytic vesicle', 'phagocytic vesicle membrane', 'U5 snRNP', 'MHC class I peptide loading complex', 'extrinsic component of organelle membrane']:
    droplist.remove(i)
df=prepare_data(found_list, droplist)
df.head(5)
```

# Draw the chart 

```python
def drawboxplot(df):
    compd=pd.DataFrame()

    for i in range(df.shape[0]):
    # for i in range(400):
        if(len(list(df['k'][i])) >=5):
            perc= pd.DataFrame()
            perc["k"]=df['k'][i]
    #         perc["complex_name"]=df['complex_name'][i] + " ("+str(i)+"  "+str(len(list(df['k'][i]))) +")"
            perc["complex_name"]=df['complex_name'][i] + " ("+str(len(list(df['k'][i]))) +")"
            compd=pd.concat([compd,perc])

    # fig = plt.gcf()
    fig, ax = plt.subplots(1)

    # Change seaborn plot size
    fig.set_size_inches(9, 100)

    sns.boxplot(x="k", y="complex_name", data=compd,orient ='h')
    # sns.swarmplot(x="k", y="name", data=compd,orient='h', color=".25")
    plt.xlabel('')
    plt.ylabel('')
    # sns.swarmplot(x="name", y="k", data=compd, color=".25")        
drawboxplot(df)
```
